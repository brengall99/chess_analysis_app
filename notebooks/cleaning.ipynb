```python
def unix_to_real_time(unix_time):
    if isinstance(unix_time, (int, float)):
        real_time = datetime.datetime.fromtimestamp(unix_time, datetime.timezone.utc)
        return real_time.strftime('%H:%M:%S')
    else:
        return pd.NaT 

df['end_time'] = df['end_time'].apply(unix_to_real_time)
```

```python
print(list(df.columns))
```

```python
df = df[['url', 
	'time_control', 
	'rated', 
	'white_username', 
	'white_rating', 
	'black_username', 
	'black_rating', 
	'result', 
	'Result',
	'Date',
	'EndDate',
	'StartTime', 
	'end_time', 
	'fen',  
	'ECO', 
	'ECOUrl', 
	'Event',  
	'Termination']]
```

```python
df = df.rename(columns={  
    	'Result': 'result_games',
    	'Date': 'start_date',
    	'EndDate': 'end_date',
    	'StartTime': 'start_time',  
    	'ECO': 'eco',
    	'ECOUrl': 'eco_url',
    	'Event': 'event',
    	'Termination': 'termination'})
```

```python
df.head()
```

```python
print((df['start_date'] != df['end_date']).sum()) # if the game passed through midnight
```

```python
df['brengall99_rating'] = df.apply(
    lambda row: row['white_rating'] if row['white_username'] == 'brengall99' else
                row['black_rating'] if row['black_username'] == 'brengall99' else None, 
    axis=1)
```

```python
df['opponent_rating'] = df.apply(
    lambda row: row['white_rating'] if row['white_username'] != 'brengall99' else
                row['black_rating'] if row['black_username'] != 'brengall99' else None, 
    axis=1)
```

```python
df = df[['url', 'time_control', 'rated', 'brengall99_rating', 'opponent_rating', 'white_username', 'white_rating',
       'black_username', 'black_rating', 'result', 'result_games',
       'start_date', 'end_date', 'start_time', 'end_time', 'fen', 'eco',
       'eco_url', 'event', 'termination']]
```
